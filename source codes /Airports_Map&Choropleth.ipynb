```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
import geoplotlib
import json
import folium
import os
from folium import plugins
from folium.plugins import *
from folium.utilities import *
from folium.features import *
import requests
from xml.etree import ElementTree
import branca.colormap as cm
```

```python
#Importing Datasets

airports= pd.read_csv('/Users/noraamri/Documents/mastars  2nd semester /visualisation /Project /data/airports_cleaned.csv').dropna()
airports = airports[["Name", "City", "Country", "latitude", "longitude"]].drop_duplicates()
airports.Name.apply(str)
airports.head()
countries_total = pd.read_csv('/Users/noraamri/Documents/mastars  2nd semester /visualisation /Project /data/countries_total1.csv')
```

```python
#importing Json file 
world_geo = os.path.join('/Users/noraamri/Documents/', 'world-countries.json')
geo_json_data = json.load(open(world_geo))

#putting airport locations in a list
locations = airports[['latitude', 'longitude']]
locationlist = locations.values.tolist()
len(locationlist)

Airport_names = airports['Name'].apply(str)
nameslist = Airport_names.values.tolist()
len(nameslist)


countries = pd.read_csv('/Users/noraamri/Documents/mastars  2nd semester /visualisation /Project /data/countries of the world 3.csv')
countries = countries[['Country','ID','Population','GDP ($ per capita)',]].drop_duplicates()
countries['Country'] = countries['Country'].astype('str') 
countries['Country'] = countries['Country'].str.strip()
countries.rename(columns={'GDP ($ per capita)': 'GDP'}, inplace= True)
countries.head()
countries.dtypes
countries.head()
```

```python
#Creating the map 
m= folium.Map(location=[airports['latitude'].mean(),airports['longitude'].mean()],
              zoom_start=3,
              tiles = 'Mapbox Bright')

m.add_tile_layer(tiles='stamenterrain', name = 'stamenterrain')

 #creating a Choropleth of GDP on the map  
folium.Choropleth(geo_data=world_geo,
                  name=' GDP Choropleth',
                  data=countries,
                  columns=['ID','GDP'],
                  key_on='feature.id',
                  fill_color='YlGn',
                  fill_opacity=0.7,
                  line_opacity=0.5,
                  legend_name='GDP ($ per capita)',
                  reset=True).add_to(m)


#Adding Airports points to the map
airports_markers = folium.FeatureGroup(name="Airports")
for point in range(0, len(locationlist)):
    airports_markers.add_child(folium.CircleMarker(location=locationlist[point],radius=1))
       

#adding the layer controller   
m.add_child(airports_markers)
m.add_child(folium.LayerControl())

m.save('Airports.html')
```
